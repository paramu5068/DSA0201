```python
# STEP 1: Install required packages (if not already installed)
!pip install opencv-python-headless matplotlib

# STEP 2: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
import io

# STEP 3: Upload the image
uploaded = files.upload()

# STEP 4: Read the uploaded image
for filename in uploaded.keys():
    image_path = filename

# Convert to grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# STEP 5: Define High-Boost Mask (adjust A value as needed)
A = 2  # You can change this value (e.g., 1 for standard Laplacian, >1 for high-boost)
high_boost_kernel = np.array([
    [-1, -1, -1],
    [-1, A + 8, -1],
    [-1, -1, -1]
])

# STEP 6: Apply filter
sharpened = cv2.filter2D(image, -1, high_boost_kernel)

# STEP 7: Display the original and sharpened images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"High-Boost Sharpened (A={A})")
plt.imshow(sharpened, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
