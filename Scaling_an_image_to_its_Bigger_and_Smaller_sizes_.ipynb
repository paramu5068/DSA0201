```python
!pip install opencv-python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
import io
import numpy as np

image_path = next(iter(uploaded))  # Get uploaded file name
image = cv2.imdecode(np.frombuffer(uploaded[image_path], np.uint8), cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for display

# Scale down (50%) and scale up (200%)
smaller = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)
bigger = cv2.resize(image, (0, 0), fx=2, fy=2)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(smaller)
plt.title('Smaller (20%)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(bigger)
plt.title('Bigger (100%)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
