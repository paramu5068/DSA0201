```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

image_path = next(iter(uploaded))
image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[100, 100], [25, 50], [10, 150]])

matrix = cv2.getAffineTransform(pts1, pts2)

transformed_image = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

transformed_image_rgb = cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

# Transformed Image
plt.subplot(1, 2, 2)
plt.imshow(transformed_image_rgb)
plt.title("Affine Transformed Image")
plt.axis('off')

plt.show()
```
