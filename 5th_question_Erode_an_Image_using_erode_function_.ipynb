```python
# Install necessary libraries
!pip install opencv-python-headless matplotlib

# Upload the image
from google.colab import files
uploaded = files.upload()

# Import necessary libraries
import cv2
import matplotlib.pyplot as plt

# Read the image
image_path = next(iter(uploaded))  # Get the uploaded image name
image = cv2.imread(image_path)

# Convert the image from BGR to RGB (for displaying with Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Convert the image to grayscale (erosion is typically applied to grayscale or binary images)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Create a kernel (a 5x5 matrix of ones)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))  # You can adjust the size

# Apply erosion
image_eroded = cv2.erode(image_gray, kernel, iterations=1)

# Display the eroded image
plt.imshow(image_eroded, cmap='gray')
plt.axis('off')
plt.show()

# (Optional) Save the eroded image
cv2.imwrite('eroded_image.jpg', image_eroded)
```
