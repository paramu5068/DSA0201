```python
# STEP 1: Install necessary packages
!pip install opencv-python-headless matplotlib

# STEP 2: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# STEP 3: Upload an image
uploaded = files.upload()
for filename in uploaded.keys():
    image_path = filename

# STEP 4: Read and convert to grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# STEP 5: Define a simple edge-detection kernel (e.g., Laplacian)
kernel = np.array([[0, -1,  0],
                   [-1,  4, -1],
                   [0, -1,  0]])

# STEP 6: Apply convolution using filter2D
boundary = cv2.filter2D(image, -1, kernel)

# STEP 7: Display original and boundary-detected image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Boundary Detected")
plt.imshow(boundary, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
