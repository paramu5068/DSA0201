```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload image
uploaded = files.upload()
image_path = next(iter(uploaded))

# Step 2: Read image and convert to grayscale
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Apply Gaussian Blur
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Step 4: Create Unsharp Mask
mask = cv2.subtract(gray, blurred)
sharpened = cv2.add(gray, mask)  # k=1, so sharpened = original + mask

# Optional: Increase sharpening effect
# sharpened = cv2.addWeighted(gray, 1.5, blurred, -0.5, 0)

# Step 5: Display original and sharpened images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image (Unsharp Masking)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
